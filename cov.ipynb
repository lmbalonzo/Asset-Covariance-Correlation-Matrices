```python
import pandas_datareader as pd 
import datetime as dt
import seaborn as sns
```

```python
end = dt.datetime.now()
start = dt.datetime(end.year-1,end.month,end.day)
```

```python
portfolio = ['FB','AMZN','AAPL','NFLX','GOOG','^GSPC','PSEI.PS','BTC-USD','ETH-USD']
```

```python
df = pd.get_data_yahoo(portfolio,start,end)['Adj Close']
returns=df.pct_change() #Percent Changes for the Past Year
returns
```

```python
returns.cov()   # Covariance Shows Direction of Relationship
```

```python
returns.var() # Variance of an asset is equal to covariance to itself
```

```python
returns.corr() # Covariance Shows Strength of Relationship
```

```python
sns.heatmap(returns.corr(), ) #Visualization of Correlation
```
